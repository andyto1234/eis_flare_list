```python
import pandas as pd
import sqlite3
import os
import numpy as np
```

```python
# Expand the user's home directory
home = os.path.expanduser("~")
db_path = os.path.join(home, "Downloads", "eis_cat.sqlite")

# Connect to the SQLite database
conn = sqlite3.connect(db_path)

# Read all tables into a dictionary of DataFrames
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'", conn)
dataframes = {}

for table_name in tables['name']:
    dataframes[table_name] = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)

# Close the connection
conn.close()

# Now you have all tables as pandas DataFrames in the 'dataframes' dictionary
# You can access them like this:
for table_name, df in dataframes.items():
    print(f"Table: {table_name}")
    print(df.head())
    print("\n")

# To work with a specific table, you can do:
# df = dataframes['your_table_name']
# Then use pandas functions on df as needed
```
