```python
from sunpy.net import Fido, attrs as a
import numpy as np
from astropy.time import Time, TimeDelta
import astropy.units as u
import sunpy.map
from scipy.ndimage import gaussian_filter


def find_flare_location(hek_entry):
    time_delta = TimeDelta(30, format='sec')
    start_time = hek_entry["event_starttime"]
    peak_time = hek_entry['event_peaktime']
    start_time_lagged = start_time + time_delta
    peak_time_lagged = peak_time + time_delta
    result_start = Fido.search(a.Time(start_time, start_time_lagged),
                         a.Instrument("aia"), a.Wavelength(94*u.angstrom))[0][0]
    result_peak = Fido.search(a.Time(peak_time, peak_time_lagged),
                         a.Instrument("aia"), a.Wavelength(94*u.angstrom))[0][0]
    downloaded_files = Fido.fetch(result_start, result_peak) 
    maps = sunpy.map.Map(downloaded_files)
    
    start_map = maps[0]/maps[0].exposure_time
    end_map = maps[1]/maps[1].exposure_time
    diff = end_map.data-start_map.data # find the difference between two maps
    
    
    
    blurred = gaussian_filter(diff, sigma=20) # apply blurring
    diff_map=sunpy.map.Map(blurred, start_map.meta)
    diff_map.peek(vmin=0,vmax=100)
    
    pixel_pos = np.argwhere(diff_map.data == diff_map.data.max()) * u.pixel
    hpc_max = diff_map.wcs.pixel_to_world(pixel_pos[:, 1], pixel_pos[:, 0])
    return hpc_max.Tx.value[0], hpc_max.Ty.value[0]
```

```python
import time
from sunpy.net import Fido, attrs as a
import pandas as pd
from datetime import datetime, timedelta

event_type = "FL"
start_year = 2011
end_year = datetime.now().year  # Current year

max_retries = 15
retry_delay = 1  # starting delay in seconds

for year in range(start_year, end_year + 1):
    tstart = f"{year}/01/01 00:00:00"
    tend = f"{year}/12/31 23:59:59"
    
    print(f"Processing year {year}")
    
    for attempt in range(max_retries):
        try:
            result = Fido.search(a.Time(tstart, tend),
                                 a.hek.EventType(event_type),
                                 a.hek.OBS.Observatory == "GOES")
            hek_results = result["hek"]
            df = hek_results.to_pandas()
            
            # Save the DataFrame to a CSV file
            csv_filename = f'{year}_hek_results.csv'
            df.to_csv(csv_filename, index=False)
            print(f"Results for {year} saved to {csv_filename}")
            break  # If successful, break out of the retry loop
        
        except Exception as e:
            print(f"Attempt {attempt + 1} for year {year} failed: {e}")
            if attempt < max_retries - 1:
                sleep_time = retry_delay * (1.5 ** attempt)  # Exponential backoff
                print(f"Retrying in {sleep_time} seconds...")
                time.sleep(sleep_time)
            else:
                print(f"Max retries reached for year {year}. Moving to next year.")
    
    # Add a delay between years to avoid overwhelming the server
    time.sleep(10)

print("Processing complete.")
```
