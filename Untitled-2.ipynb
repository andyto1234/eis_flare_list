```python
import re

def parse_raster_entry(entry):
    # Extract raster name (looking for the longest sequence of printable characters before [RA:])
    name_match = re.search(r'([A-Za-z0-9_+\-\.]+(?:[_+\-\. ][A-Za-z0-9_+\-\.]+)*)\s*\[RA:', entry)
    name = name_match.group(1).strip() if name_match else ''
    
    # If name is empty, try to find it after [RA:]
    if not name:
        name_match = re.search(r'\[RA:\d+\]\s*([A-Za-z0-9_+\-\.]+(?:[_+\-\. ][A-Za-z0-9_+\-\.]+)*)', entry)
        name = name_match.group(1).strip() if name_match else ''
    
    # Extract other information using regular expressions
    ra_number = re.search(r'\[RA:(\d+)\]', entry)
    scan_info = re.search(r'\[Scan:(\d+)\((\d+)"?\)steps\]', entry)
    sns_info = re.search(r'\[SNS:(\d+)sets\]', entry)
    ss_info = re.search(r'\[ss:(\d+)"\]', entry)
    wh_info = re.search(r'\[wH:(\d+)', entry)
    exp_info = re.search(r'\[ExpT\(Delay\):(\d+\.\d+)', entry)
    author_date = re.search(r'(\w+\s+\w+)\s+(\d+-\w+-\d+)', entry)

    return {
        'name': name,
        'ra_number': ra_number.group(1) if ra_number else None,
        'scan_steps': scan_info.group(1) if scan_info else None,
        'scan_size': scan_info.group(2) if scan_info else None,
        'sns_sets': sns_info.group(1) if sns_info else None,
        'ss': ss_info.group(1) if ss_info else None,
        'wh': wh_info.group(1) if wh_info else None,
        'exp_time': exp_info.group(1) if exp_info else None,
        'author': author_date.group(1) if author_date else None,
        'date': author_date.group(2) if author_date else None
    }

# Assuming you've already read the content of the file into the 'content' variable
raster_entries = content.split('\n')
rasters = [parse_raster_entry(entry) for entry in raster_entries if entry.strip()]

# Filter out entries with empty names
rasters = [raster for raster in rasters if raster['name']]

# Now you have a list of dictionaries, each containing information about a raster
for raster in rasters:
    print(raster)
```

```python
def get_total_exposure_time(name):
    for raster in rasters:
        if raster['name'] == name:
            exp_time = float(raster['exp_time']) if raster['exp_time'] else 0
            scan_steps = int(raster['scan_steps']) if raster['scan_steps'] else 0
            return exp_time * scan_steps
    return None  # Return None if the name is not found

# Example usage:
# name = "example_raster_name"
# total_exposure_time = get_total_exposure_time(name)
# if total_exposure_time is not None:
#     print(f"Total exposure time for {name}: {total_exposure_time}")
# else:
#     print(f"Raster {name} not found")
```
